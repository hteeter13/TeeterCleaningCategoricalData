```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
display(customer_df.shape)
customer_df.head(10)
```

```python
customer_df.dtypes
```

```python
cols = []
for c in customer_df.columns:
    cols.append(c.lower())
customer_df.columns = cols
```

```python
customer_df.info()
```

```python
customer_df.describe().T
```

```python
customer_df.columns = customer_df.columns.str.lower()
customer_df
```

```python
customer_df['effective to date'] = pd.to_datetime(customer_df['effective to date'])
customer_df
```

```python
categorical_df = customer_df.select_dtypes(object)
categorical_df.head(10)
```

```python
categorical_df.isna().sum()/len(categorical_df)
```

```python
categorical_df.shape
```

```python
categorical_df = categorical_df.drop(['customer'], axis = 1)
```

```python
def plot_cat(Dataframe):
    for column in Dataframe.columns:
        sns.countplot(x=Dataframe[column])
        plt.show()
    return None
   

plot_cat(categorical_df)
```

```python
categorical_df_columns = [col for col in categorical_df if (len(categorical_df[col].unique()) < (customer_df.shape[0] * 0.01))]
```

```python
print(categorical_df_columns)
```

```python
categorical_df.head()
```

```python
categorical_df["coverage"] = categorical_df["coverage"].map({"Basic" : 0, "Extended" :.5, "Premium" : 1})
```

```python
categorical_df['employmentstatus'].unique()
```

```python
categorical_df["employmentstatus"] = categorical_df["employmentstatus"].map({"Medical Leave" : 0, "Disabled" : 0, "Unemployed" : 0, "Retired" : .5, "Employed":2})
```

```python
categorical_df["education"] =categorical_df["education"].map({"High School or Below" : 0, "Bachelor" : .5, "College":.5, "Master" : 1, "Doctor":1})
```

```python
categorical_df["location code"] =categorical_df["location code"].map({"Rural" : 0, "Suburban" :.5, "Urban" : 1,})
```

```python
categorical_df["vehicle size"] =categorical_df["vehicle size"].map({"Small" : 0, "Medsize" :.5, "Large" : 1,})
```
